## Q. 91-92. k-meansによる減色
チュートリアルにまんま載ってるので気も引けるが書いておく。RGBの3次元空間の特徴量と考えるので，RGBの各チャネルを1次元化してからk-meansを実行する。RGBを量子化したい場合には前の問題でやった減色を使えばよい。

```python
import numpy as np
img = cv2.imread('Calc.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
Z = img.reshape((-1, 3)).astype(np.float32)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 1.0)
ret, labels, centers = cv2.kmeans(Z, 5, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
res = centers[labels.flatten()]
res2 = res.reshape((img.shape))
plt.figure()
plt.imshow(cv2.cvtColor(res2, cv2.COLOR_BGR2RGB))
```
